## Filtro zonas eleitorais do muniícpio do Rio

Esse notebook lê o arquivo com lista de zonas eleitorais do Estado do Rio fornecido pelo TRE-RJ, originalmente em XLS, e gera um novo arquivo *.csv*. A partir desse novo arquivo, filtra apenas as zonas eleitorais do município do Rio.

O arquivo .xls original pode ser encontrado no site do TRE-RJ em:
http://www.tre-rj.jus.br/site/eleicoes/2016/arquivos/porBairroZon.xls

```python
import xlrd
import csv
import pandas as pd

# Abre o arquivo .xls e converte para .csv
# Nesse procedimento, os números serão gravados como float

with xlrd.open_workbook('src/porBairroZon.xls') as wb:
    sh = wb.sheet_by_index(0)            
    with open('src/porBairroZon_v3.csv', 'w', newline="") as f:
        c = csv.writer(f)
        for r in range(sh.nrows):
            c.writerow(sh.row_values(r))

# Abre o novo arquivo .csv com pandas
lista_zonas = pd.read_csv("src/porBairroZon_v3.csv", encoding='latin-1')
zonas = pd.DataFrame(lista_zonas)

zonas.head()
```

```python
zonas.sort_values(by="Municipio")
```
